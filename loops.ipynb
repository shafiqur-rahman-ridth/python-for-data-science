## While Loop
* if-elif-else construct, a way to alter the flow of your scripts
* as soon as python encounters an if statement, it checks the condition. When the condtion is true, the corresponding code is executed. If the condition evaluates to False, and if there are elif and else statement, Python continues his search. Anyways python goes throught this code only once. After that, Python heads over to the next command in the script.
* while loop is less used. But in some cases, it can be very helpful.
* the syntax of the while loop is very similar to the if statement

* It executes the code inside if the condition is true. However, as opposed to the if statement.

* **while loop = repeated if statement**
* **or repeating action untill condition is met**
* **as long as the condition is true, the while loop will continue to execute his code over and over again**
* **while loop stops when the condition is fale**

```python
while condition:
    expression
```

* Keep going untill the error is no longer above 1

```python
error = 50 
# While loop executes as long as the error variable is above 1
# if the error variable is greater than 1 (condtion), keep running the code over and over again
while error > 1:
    print(error)
    error = error/4 
    print(error)
```

* First run --> error is 50, condition is true (error >1) , execute the code
* Second run --> error is 12.5, condition is true (error > 1), execute the code
* Third run --> error is 3.125, codition is true (error > 1), execute the code
* Fourth run --> error is 0.78125, condition is false (error < 1), stop while loop code                                        

### Forever while loop
* So be careful, your while loop shall stop some time

```python
error = 50
while error > 1:
    print(error)
```

# For Loop

## for loop syntax
* **for each variable (var/element), in a sequence (seq/list/df/dictionary), execute the expression**

```python
for var in seq:
    expression
```

## lists/pandas series: for loop
* print out each element in a list seperately

```python
fam = [1.73, 1.68, 1.71, 1.89]
print(fam[0])
print(fam[1])
print(fam[2])
print(fam[3])
```

* Instead of the above repetitive and rediuos approach, you shall use for loop
* Execute some code for each height in the fam list
* I want to do something (expression) on each element (give any arbirary name) from the list (name of original list)
* Process continues untill all heights in fam list have beem iterated over

```python
for height in fam:
    print(height)
```

```python
type(height)
```

## for loop: list with index
* use enumerate()
* this time iterate/loop over each index (element1) and height (element2) from list fam

```python
for index, height in enumerate(fam):
    print('index_value', index, ':', height)
```

* Converted the floats and other data types into one types string while printing using plus (+)

```python
for index, height in enumerate(fam):
    print('index ' + str(index) + ": " + str(height))
```

## Loop over: strings
* iterates over every character in a string

```python
for c in 'family':
    print(c.capitalize())
```

## Dictionary: Loop Over
* what about loop over other data structures, such as dictionaries and Numpy array
* In both cases, we can use similar for loop construct, but the way you define the "sequence" over which you are iterating will differ depending in the data structure
* use methods items()
* dictionaries are inherently ordered: the order in which they are iterated over is not fixed
* key, values are arbitrary values
* the first variable gets key and second one the value

```python
world = { "afghanistan":30.55,
"albania":2.77,
"algeria":39.21 }

for key, value in world.items():
    print(key + " -- " + str(value))
```

## Two-Dimensional Numpy Array: Loop Over
* Same loop over for 1-D numpy array won't work
* To visit every element of an array, use nditer() function

```python
import numpy as np
np_height = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
np_weight = np.array([65.4, 59.2, 63.6, 88.4, 68.7])
two_dimensional_numpyarray = np.array([np_height, np_weight])
```

```python
two_dimensional_numpyarray
```

* Visiting each list, but not each element

```python
for val in two_dimensional_numpyarray:
    print(val)
```

* to iterate each element use np.nditer() function on 2d numpy array 
* printing all the height and weights element-wise

```python
for val in np.nditer(two_dimensional_numpyarray):
    print(val)
```

## Pandas DataFrame: Loop Over
* loop over df means I have all rows available to be used

```python
import pandas as pd 
brics = pd.read_csv('brics.csv', index_col=0)
brics
```

```python
type(brics)
```

* if a pandas dataframe were to function the same way as a 2D Numpy array, may be a basic for loop like this, to print out each row could work

* this prints only column names of DataFrame, this is not exactly we want

```python
for val in brics:
    print(val)
```

* In Pandas, mention explicit that you want to iterate over the rows 
* **iterrows() --> iterate each rows in a dataframe**
* use iterrows() method on the DF
* iterrows() method looks at the df, and on each iteration **generates two pieces of data: the label of row and secondly actual data in the row as a Pandas Series**
* store row label as lab, and row data as row. Thus, lab and row are arbitrary, any names can be given

```python
brics 
```

* Because this row variable on each iteration is a Series, you can easily select additional information from it using subsetting techniques.

```python
for lab, row in brics.iterrows():
    print(lab)
    print(row)
```

* Suppose you only want to print out the capital in each iteration
* change print statement as follows:

```python
for lab, row in brics.iterrows():
    print(lab + ": " + row['capital'])
```

```python
for lab, row in brics.iterrows():
    print(lab, row['area'])
```

```python
for lab, row in brics.iterrows():
    print(lab, row['capital'], row['population'])
```

## Add column: DF

* Not efficient, because you are creating a Series object on every iteration

```python
# All rows are available to us when we loop over df
for lab, row in brics.iterrows():
    brics.loc[lab, "name_length"] = len(row['country'])
print(brics)
```

## apply() function
* use for loop, creates a Pandas Series object on every iteration. Thus, it is inefficient process.
* Way better approach, if you want to calculate an entire DataFrame column by applying a function on a particular column in an element-wise fashion. 
* do vectorized operation with apply() function
* **then use apply()**

## use of len function on a specific df column by apply function 
* when using len() function inside the apply function, no need to specify parenthesis or first bracket ()

```python
brics["name_length_apply"] = brics['country'].apply(len)
```

## use of upper function on a specific df column by apply function 

```python
brics["name_length_apply_2"] = brics['country'].apply(str.upper)
```
